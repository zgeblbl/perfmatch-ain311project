```python
# load the data
team_df_original= pd.read_parquet('parquet_data/team_df_missing_handled.parquet')
player_df_original= pd.read_parquet('parquet_data/player_df_missing_handled.parquet')

# make a success colon to be predicted
team_df['Success'] = team_df_original['Result'].apply(
    lambda x: 1 if isinstance(x, str) and x.startswith('W') else (0 if isinstance(x, str) and x.startswith('L') else None))
player_df['Success'] = player_df_original['Result'].apply(
    lambda x: 1 if isinstance(x, str) and x.startswith('W') else (0 if isinstance(x, str) and x.startswith('L') else None))
```

### Data Balancing Techniques
Oversampling: To increase the number of samples for the underrepresented class, methods such as SMOTE (Synthetic Minority Over-sampling Technique) or Random Over-sampling can be used. This helps balance the dataset by artificially increasing the data for the rare class.

Undersampling: This involves randomly removing some samples from the majority class. It is used to balance the dataset, but this method can lead to data loss.
